# Decision Trees
**Chapter 6 — Hands-On ML**

Covers: classification tree (Iris), tree visualization via Graphviz, regression tree.

Dataset: Iris — petal length & width features only.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, export_graphviz
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')
```

## 1. Dataset

```python
iris = datasets.load_iris()
X = iris['data'][:, 2:]    # petal length, petal width
y = iris['target']
print('Classes:', iris.target_names)
print('Feature names:', iris.feature_names[2:])
```

## 2. Decision Tree Classifier

```python
tree_clf = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clf.fit(X, y)
print('Depth:', tree_clf.get_depth())
print('Leaves:', tree_clf.get_n_leaves())
```

## 3. Tree Visualization

Exports a `.dot` file. Render with: `dot -Tpng iris_tree.dot -o iris_tree.png`

```python
IMAGES_PATH = Path('.')
export_graphviz(
    tree_clf,
    out_file=str(IMAGES_PATH / 'iris_tree.dot'),
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True,
)
print('Exported iris_tree.dot — run: dot -Tpng iris_tree.dot -o iris_tree.png')
```

## 4. Predictions

Each leaf stores the class probabilities. Prediction = class with highest probability at the leaf reached by the sample.

```python
sample = [[5, 1.5]]
print('Predicted class:', iris.target_names[tree_clf.predict(sample)[0]])
print('Class probabilities:', tree_clf.predict_proba(sample))
```

## 5. Decision Boundary Visualization

```python
x0, x1 = np.meshgrid(np.linspace(0, 8, 500), np.linspace(0, 4, 200))
X_grid = np.c_[x0.ravel(), x1.ravel()]
y_grid = tree_clf.predict(X_grid).reshape(x0.shape)

plt.figure(figsize=(9, 4))
plt.contourf(x0, x1, y_grid, cmap=plt.cm.brg, alpha=0.2)
for cls, name, marker in zip(range(3), iris.target_names, ['bs', 'g^', 'ro']):
    plt.plot(X[y == cls, 0], X[y == cls, 1], marker, label=name, markersize=6)
plt.xlabel('Petal Length (cm)'); plt.ylabel('Petal Width (cm)')
plt.title('Decision Tree — Decision Boundaries (max_depth=2)')
plt.legend(); plt.show()
```

## 6. Decision Tree Regression

Instead of class labels, leaves store the *mean target value* of training samples that reached them.

```python
tree_reg = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg.fit(X, y)
print('Regression tree predictions (first 5):', tree_reg.predict(X[:5]))
print('Actual targets (first 5):', y[:5])
```
