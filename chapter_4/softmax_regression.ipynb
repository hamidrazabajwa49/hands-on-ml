# Softmax Regression
**Chapter 4 — Hands-On ML**

Extends logistic regression to multiclass — learns K sets of weights, outputs class probabilities via softmax.

Dataset: Iris — 3-class classification using petal length & width.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
```

## 1. Dataset

```python
iris = datasets.load_iris()
X = iris['data'][:, (2, 3)]   # petal length, petal width
y = iris['target']             # 0=setosa, 1=versicolor, 2=virginica
print('X shape:', X.shape)
print('Classes:', iris.target_names)
```

## 2. Train Softmax Classifier

Sklearn's `LogisticRegression` uses softmax automatically when `multi_class='multinomial'`.

```python
softmax_reg = LogisticRegression(solver='lbfgs', C=10, random_state=42)
softmax_reg.fit(X, y)
```

## 3. Predict & Probability

```python
sample = [[5, 2]]
print('Predicted class:', softmax_reg.predict(sample))
print('Predicted class name:', iris.target_names[softmax_reg.predict(sample)[0]])
print('\nClass probabilities:')
for name, prob in zip(iris.target_names, softmax_reg.predict_proba(sample)[0]):
    print(f'  {name}: {prob:.4f}')
```

## 4. Decision Boundary Visualization

```python
x0, x1 = np.meshgrid(np.linspace(0, 8, 500), np.linspace(0, 3.5, 200))
X_grid = np.c_[x0.ravel(), x1.ravel()]
y_grid = softmax_reg.predict(X_grid).reshape(x0.shape)

plt.figure(figsize=(10, 4))
plt.contourf(x0, x1, y_grid, cmap=plt.cm.brg, alpha=0.2)
for cls, name, marker in zip(range(3), iris.target_names, ['bs', 'g^', 'ro']):
    plt.plot(X[y == cls, 0], X[y == cls, 1], marker, label=name, markersize=6)
plt.xlabel('Petal Length (cm)'); plt.ylabel('Petal Width (cm)')
plt.title('Softmax Regression — Decision Boundaries')
plt.legend(); plt.show()
```
