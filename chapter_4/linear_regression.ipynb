# Linear Regression
**Chapter 4 — Hands-On ML**

Covers Normal Equation, Batch GD, Stochastic GD (manual + sklearn).

Dataset: Synthetic — `y = 4 + 5x + noise`

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, SGDRegressor
import warnings
warnings.filterwarnings('ignore')
```

## 1. Dataset

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 5 * X + np.random.randn(100, 1)

plt.scatter(X, y, alpha=0.6)
plt.xlabel('X'); plt.ylabel('y'); plt.title('Synthetic Dataset')
plt.show()
```

## 2. Normal Equation
$$\hat{\theta} = (X^T X)^{-1} X^T y$$

```python
X_b = np.c_[np.ones((100, 1)), X]
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
print('theta (intercept, slope):', theta_best.ravel())
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_pred = X_new_b @ theta_best

plt.plot(X_new, y_pred, 'r-', linewidth=2, label='Prediction')
plt.scatter(X, y, alpha=0.4, label='Data')
plt.axis([0, 2, 0, 15]); plt.legend(); plt.title('Normal Equation — Fit')
plt.show()
```

## 3. Sklearn LinearRegression

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
print('Intercept:', lin_reg.intercept_)
print('Coefficient:', lin_reg.coef_)
```

## 4. Batch Gradient Descent
$$\theta \leftarrow \theta - \eta \frac{2}{m} X^T (X\theta - y)$$

```python
eta, n_iterations, m = 0.1, 1000, 100
theta = np.random.randn(2, 1)
for _ in range(n_iterations):
    gradients = (2 / m) * X_b.T @ (X_b @ theta - y)
    theta -= eta * gradients
print('theta (BGD):', theta.ravel())
```

## 5. Stochastic Gradient Descent (Manual)

```python
n_epochs, t0, t1 = 50, 5, 50

def learning_schedule(t):
    return t0 / (t + t1)

theta = np.random.randn(2, 1)
for epoch in range(n_epochs):
    for i in range(m):
        idx = np.random.randint(m)
        xi, yi = X_b[idx:idx+1], y[idx:idx+1]
        gradients = 2 * xi.T @ (xi @ theta - yi)
        eta = learning_schedule(epoch * m + i)
        theta -= eta * gradients
print('theta (SGD manual):', theta.ravel())
```

## 6. SGD — Sklearn

```python
sgd_reg = SGDRegressor(max_iter=50, penalty=None, eta0=0.1, random_state=42)
sgd_reg.fit(X, y.ravel())
print('Intercept:', sgd_reg.intercept_)
print('Coefficient:', sgd_reg.coef_)
```
